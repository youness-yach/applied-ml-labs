for loops

```python
#loop over elements of a list and compute their sum

total = 0 
for value in [3, 2, 19]: 
    total += value
print('total: {0}'.format(total))
```

```python
#can be nested

total = 0
for value in [2, 3, 19]:
    for weight in [3, 2, 1]:
        total += value * weight
print('total: {0}'.format(total)) #to sum each combination of value and weight 
```

```python
#to sum over value, weight pairs 
#-- to compute the avg value of a random variable that takes on possible values 2,3,19
#with probability 0..2,0.3,0.5 respectively
#using zip() function: loops over a sequence of tuples

total = 0
for value, weight in zip([2,3,19],[0.2,0.3,0.5]):
    total += value * weight
print('weighted average: {0}'.format(total))
```
